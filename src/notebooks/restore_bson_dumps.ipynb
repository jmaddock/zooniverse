```python
import pandas as pd
import os
from pymongo import MongoClient
import subprocess
import shutil
```

```python
ignore_file_list = [
    'favorites.bson',
    'projects.bson',
    'talk.bson',
    'users.bson',
    'boards.bson'
]
base_dir = r'/srv/zooniverse/raw_data/ouroboros/ouroboros_2016-08-11'

infile_list = [x for x in os.listdir(base_dir) if x not in ignore_file_list]
for infile in infile_list:
    collection_name = infile.replace('.bson','')
    infile_path = os.path.join(base_dir,infile)
    print('importing {0} to {1}'.format(infile_path,collection_name))
    subprocess.run(['mongorestore','-d','zooniverse','-c',collection_name,infile_path])
    break
```

```python
fieldfile = r'/home/sjm668/dev/zooniverse/src/classification_fields.txt'
outdir = r'/srv/zooniverse/tables'

client = MongoClient()
collections_list = client['zooniverse'].collection_names()
tempdir = os.path.join(outdir,'temp')
if not os.path.exists(tempdir):
    os.mkdir(tempdir)
for collection in collections_list:
    temp_outfile = os.path.join(tempdir,'{0}.csv'.format(collection))
    print('exporting {0} to {1}'.format(collection,temp_outfile))
    subprocess.run(['mongoexport','-d','zooniverse','-c',collection,'--fieldFile',fieldfile,'--type=csv','--out',temp_outfile])
    break
```
